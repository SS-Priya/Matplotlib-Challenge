```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame

combined_mouse_df=pd.merge(study_results,mouse_metadata,on='Mouse ID',how='left')
# Display the data table for preview
combined_mouse_df.head()
```

```python
# Checking the number of mice.
combined_mouse_df['Mouse ID'].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicate_data = combined_mouse_df[combined_mouse_df.duplicated(subset=['Mouse ID', 'Timepoint'],\
                                                                keep=False)]
duplicate_data['Mouse ID'].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

duplicate_indexs=combined_mouse_df.loc[combined_mouse_df['Mouse ID']=='g989',:]
duplicate_indexs
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

clean_mouse_df=combined_mouse_df.drop(duplicate_indexs.index)
clean_mouse_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.

clean_mouse_df['Mouse ID'].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
#and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

mean_tumor=clean_mouse_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].mean()

median_tumor=clean_mouse_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].median().round(decimals=6)

variance_tumor=clean_mouse_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].var()

std_dev_tumor=clean_mouse_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].std()

std_err_tumor=clean_mouse_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].sem()
```

```python
# Assemble the resulting series into a single summary DataFrame.

summary_statistics=pd.DataFrame({'Mean Tumor Volume':mean_tumor,
                                 'Median Tumor Volume':median_tumor,
                                 'Tumor Volume Variance':variance_tumor,
                                 'Tumor Volume Std. Dev.':std_dev_tumor,
                                 'Tumor Volume Std. Err.':std_err_tumor}).reset_index()
summary_statistics=summary_statistics.set_index("Drug Regimen")
```

```python
summary_statistics
```

```python
# Using the aggregation method, produce the same summary statistics in a single line

summary_stat_agg = clean_mouse_df.groupby('Drug Regimen').aggregate({'Tumor Volume (mm3)':['mean',\
                                                    'median','var','std','sem']})
summary_stat_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each

max_tp=pd.DataFrame(clean_mouse_df.groupby('Mouse ID')['Timepoint'].max()).reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint

max_timepoint_df=max_tp.merge(clean_mouse_df,on=["Mouse ID", "Timepoint"],how='left')

# Put treatments into a list for for loop (and later for plot labels)

regimen_list=['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

tumor_value_df=[]

tumor_value_df=max_timepoint_df[max_timepoint_df['Drug Regimen'].isin(regimen_list)]

tumor_value_df.head()
```

```python
# Create empty list to fill with tumor vol data (for plotting)
tumor_list=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for regimen in regimen_list:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    
    tumors=tumor_value_df[tumor_value_df['Drug Regimen']==regimen]['Tumor Volume (mm3)']
    # add subset 
    
    tumor_list.append(tumors)
    
    # Determine outliers using upper and lower bounds
    
    quartiles = tumors.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outlier = tumors.loc[(tumors< lower_bound)| (tumors> upper_bound)]
    
    print(regimen,outlier)
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
flierprops = dict(marker='o', markerfacecolor='r', markersize=12,
                  linestyle='none')

plt.boxplot(tumor_list,labels=regimen_list,flierprops=flierprops)
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin

l509_df=clean_mouse_df[clean_mouse_df["Mouse ID"]=='l509']
x=l509_df["Timepoint"]
y=l509_df["Tumor Volume (mm3)"]
plt.plot(x,y)
plt.title("Capomulin treatment of mouse l509")
plt.xlabel("Timepoint(Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

capomulin_df=clean_mouse_df[clean_mouse_df["Drug Regimen"]=="Capomulin"]
agg_capomulin_df=capomulin_df.groupby('Mouse ID')[['Weight (g)','Tumor Volume (mm3)']].mean()
x=agg_capomulin_df['Weight (g)']
y=agg_capomulin_df['Tumor Volume (mm3)']

plt.scatter(x,y)
plt.xlabel("Weight(g)")
plt.ylabel("Average Tumor Volume (mm3)")
                             
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

slope, intercept, r_value, p_value, std_err = st.linregress(x,y)

line = slope*x+intercept
plt.plot(x, line, 'r', label='y={:.2f}x+{:.2f}'.format(slope,intercept))

plt.scatter(x,y)
plt.xlabel("Weight(g)")
plt.ylabel("Average Tumor Volume (mm3)")
                             
print("The correlation between mouse weight and the average tumor volume is ",round(r_value,2))
plt.show()
```
